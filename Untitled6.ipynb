```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from docx import Document
from docx.shared import Inches
```

```python
tickers = ['KNCA11.SA', 'VALE3.SA', 'PETR4.SA', 'ITUB4.SA', 'MGLU3.SA', 'WEGE3.SA']
data = yf.download(tickers, start='2018-01-01', end='2025-01-01')['Adj Close'].dropna()
data.head()
```

```python
data = yf.download(tickers, start='2018-01-01', end='2025-01-01')['Close'].dropna()
```

```python
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(data[ticker], label=ticker)
plt.title('Evolução dos Preços Ajustados')
plt.legend()
plt.savefig('precos.png')
plt.show()
```

```python
data_pct = data.pct_change().dropna()
stats = data_pct.describe().T[['mean', '50%', 'std', 'max', 'min']]
stats.columns = ['Média', 'Mediana', 'Desvio Padrão', 'Máximo', 'Mínimo']
stats = stats * 100
stats
```

```python
# @title Média

from matplotlib import pyplot as plt
stats['Média'].plot(kind='hist', bins=20, title='Média')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
resultados = {}
mse_lr_dict = {}
mse_nn_dict = {}

for ticker in tickers:
    print(f"\nProcessando: {ticker}")
    serie = data[ticker].pct_change().dropna()

    X = serie[:-1].values.reshape(-1,1)
    y = serie[1:].values

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    # Regressão Linear
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    mse_lr = mean_squared_error(y_test, y_pred)
    mse_lr_dict[ticker] = mse_lr

    # Rede Neural
    model = Sequential()
    model.add(Dense(64, input_dim=1, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=50, verbose=0)

    y_pred_nn = model.predict(X_test).flatten()
    mse_nn = mean_squared_error(y_test, y_pred_nn)
    mse_nn_dict[ticker] = mse_nn

    # Gráfico comparativo
    plt.figure(figsize=(10, 4))
    plt.plot(y_test, label='Real')
    plt.plot(y_pred, label='Regressão Linear')
    plt.plot(y_pred_nn, label='Rede Neural')
    plt.title(f'Comparação dos Modelos - {ticker}')
    plt.legend()
    plt.savefig(f'comparacao_{ticker}.png')
    plt.show()

    # Armazenar resultados
    resultados[ticker] = {
        'mse_lr': mse_lr,
        'mse_nn': mse_nn
    }
```

```python
doc = Document()
doc.add_heading('Predição de Preços de Ativos Financeiros', 0)

# Gráfico de preços
doc.add_heading('Gráfico: Evolução dos Preços Ajustados', level=1)
doc.add_picture('precos.png', width=Inches(6))

# Estatísticas descritivas
doc.add_heading('Tabela: Estatísticas Descritivas dos Retornos (%)', level=1)
table = doc.add_table(rows=1, cols=len(stats.columns)+1)
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Ativo'
for i, col in enumerate(stats.columns):
    hdr_cells[i+1].text = col

for idx, row in stats.iterrows():
    row_cells = table.add_row().cells
    row_cells[0].text = idx
    for i, val in enumerate(row):
        row_cells[i+1].text = f"{val:.2f}"

# Resultados por ativo
for ticker in tickers:
    doc.add_heading(f'Análise: {ticker}', level=1)

    # Gráfico comparativo
    doc.add_heading('Gráfico: Comparação dos Modelos', level=2)
    doc.add_picture(f'comparacao_{ticker}.png', width=Inches(6))

    # Resultados dos modelos
    doc.add_heading('Resultados dos Modelos', level=2)
    doc.add_paragraph(f"Erro quadrático médio - Regressão Linear: {resultados[ticker]['mse_lr']:.6f}")
    doc.add_paragraph(f"Erro quadrático médio - Rede Neural: {resultados[ticker]['mse_nn']:.6f}")

# Salvar o relatório
doc.save('relatorio_completo.docx')

print("Relatório gerado: relatorio_completo.docx")
```
